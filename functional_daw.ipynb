```python
import numpy as np
import IPython.display as ipd   # using ipd.Audio and ipd.Display to play sounds
```

```python
stretch = lambda x,F : np.outer(F,x)
shift = lambda x,R : x - R[:,np.newaxis]
wav = lambda x : np.piecewise(x,[x%1 <= 0.5], [-1,1])
env = lambda x,L : np.piecewise(x, [x < 0, x > L[:,np.newaxis]], [0,0,1])
merge = lambda x,y : np.sum(x*y,axis=0)
main = lambda x,F,R,L : merge(env(shift(x,R),L),wav(stretch(x,F)))
```

```python
fs = 48000

a1 = [120,180,150,180]
a2 = [240,180,150,135]
a3 = [120,180,150,135]

b1 = [120,180,144,180]
b2 = [240,180,144,135]
b3 = [120,180,144,135]

a = np.array(a1 + a2 + a3 + a2)
b = np.array(b1 + b2 + b3 + a2)

F = 4*np.concatenate((a,3/4*a,5/6*b,5/8*a,  2/3*a,1/2*a,2/3*a,3/4*a))
num_notes = F.shape[0]

nps = 10
L = np.ones(num_notes)/nps
R = np.arange(0,num_notes)/nps

end = max(R+L)
x = np.arange(0,end,1/fs)

out = main(x,F,R,L)
out = np.tile(out,2)
```

```python
ipd.display(ipd.Audio(out, rate=fs))  
```
